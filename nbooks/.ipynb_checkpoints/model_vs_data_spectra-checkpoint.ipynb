# Compare measured data with simulated data

```python
import os
import sys
import h5py
import numpy as np
import pandas as pd
from scipy.interpolate import interp2d
from ipywidgets import interact
import matplotlib.pyplot as plt
from pyuvdata import UVData
```

## Load Visibility Data J0136
Calibrated visibilities from IDR2.1 at LST=1.6 h. Select visibilities of an EQ29 triad (37, 39, 68).

```python
vis_file = "/lustre/aoc/projects/hera/H1C_IDR3/IDR3_2/LSTBIN/all_epochs_preprocessed/zen.grp1.of1.LST.0.40170.sum.LPXL.uvh5"

# visibility data
UV = UVData()
UV.read(vis_file)

UVlst = np.unique(UV.lst_array * 12 / np.pi)
UVfrq = UV.freq_array[0] * 1e-6
#UV.select(bls=[(0, 11), (11, 12), (12, 0)])
print(np.argmin(np.abs(UVlst - 1.6)))

# (ant1, ant2, pol)
vis1 = UV.get_data(37, 39, 'xx')[[11, 12]]
vis2 = UV.get_data(39, 68, 'xx')[[11, 12]]
vis3 = UV.get_data(68, 37, 'xx')[[11, 12]]
```

## Load Simulated Data J0136

```python
trclass = "EQ28"
field = "J0136"

vis_gleam_path = f"/users/pkeller/code/ClosureSim/data/vis_{trclass}_{field}.h5"
vis_bright_path = f"/users/pkeller/code/ClosureSim/data/vis_bright_{trclass}_{field}.h5"

with h5py.File(vis_gleam_path) as f:
    lstm = f["LST"][()]
    freqm = f["Frequencies"][()]
    vis = f["Visibilities"][()]

with h5py.File(vis_bright_path) as f:
    vis += f["Visibilities"][()]

# pseudo Stokes I
vis = 0.5 * (vis[:, 0, 0] + vis[:, 1, 0])
```

## Load Validation Visibilities

```python
validation_file = "/lustre/aoc/projects/hera/Validation/test-4.0.0/pipeline/LSTBIN/foregrounds/zen.grp1.of1.LST.0.37586.HH.OCRSLP.uvh5"

# visibility data
UV = UVData()
UV.read(validation_file)

UVlst = np.unique(UV.lst_array * 12 / np.pi)
UVfrq = UV.freq_array[0] * 1e-6
#UV.select(bls=[(0, 11), (11, 12), (12, 0)])
print(np.argmin(np.abs(UVlst - 1.6)))

# (ant1, ant2, pol)
vis1v = UV.get_data(37, 39, 'xx')[[27, 28]]
vis2v = UV.get_data(39, 68, 'xx')[[27, 28]]
vis3v = UV.get_data(68, 37, 'xx')[[27, 28]]
```

## Plot Data
The model and the data seem to agree well up to some oscillations of period ~10 MHz.

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 5), sharex=True, sharey=True)

# sim
ax.plot(UVfrq, 2*np.abs(vis1v[1]), label="(validation 1)*2")
ax.plot(UVfrq, 2*np.abs(vis2v[1]), label="(validation 2)*2")
ax.plot(UVfrq, 2*np.abs(vis3v[1]), label="(validation 3)*2")

# data
ax.plot(UVfrq, np.abs(vis1[1]), alpha=0.2, label="data 1")
ax.plot(UVfrq, np.abs(vis2[1]), alpha=0.2, label="data 2")
ax.plot(UVfrq, np.abs(vis3[1]), alpha=0.2, label="data 3")

ax.legend()

plt.setp(ax, title="J0136 - 28m Baselines", xlabel="Frequency [MHz]", ylabel="Visibility Amplitude [Jy]", xlim=[110, 190], ylim=[0, 120]);
```

```python
# sim
plt.plot(UVfrq, np.abs(vis[0]), linewidth=2.0, label="my sim 1")
plt.plot(UVfrq, np.abs(vis[1]), linewidth=2.0, label="my sim 2")
plt.plot(UVfrq, np.abs(vis[2]), linewidth=2.0, label="my sim 3")

# data
plt.plot(UVfrq, np.abs(vis1[1]), alpha=0.3, label="data 1")
plt.plot(UVfrq, np.abs(vis2[1]), alpha=0.3, label="data 2")
plt.plot(UVfrq, np.abs(vis3[1]), alpha=0.3, label="data 3")

plt.legend()
plt.title("J0136 - 28m Baselines")
plt.xlabel("Frequency [MHz]")
plt.ylabel("Visibility Amplitude [Jy]")
plt.xlim([110, 190])
plt.ylim([0, 80])
```

## Noise
The noise estimated from auto-correlation visibilities and that estimated from differencing visibilities neighbouring in time seem to agree well.

```python
std_path = "/users/pkeller/code/H1C_IDR3.2/data/noise_std.h5"

# Noise estimated from auto-correlation visibilities
with h5py.File(std_path) as f:
    lstn = f["LST"][()]
    sig = f["RMS"][()].real
    
# interpolate NaN-values
lst_idx = np.argmin(np.abs(lstn - 1.6))
lstn = lstn[lst_idx]
sig = sig[lst_idx]

# data noise estimation
idx = np.where((UVfrq > 152.25) & (UVfrq < 167.97))

dvis1 = (vis1[1] - vis1[0])[idx] / np.sqrt(2)
dvis2 = (vis2[1] - vis2[0])[idx] / np.sqrt(2)
dvis3 = (vis3[1] - vis3[0])[idx] / np.sqrt(2)
```

```python
dvis = dvis3
plt.scatter(UVfrq[idx], dvis.real)
plt.scatter(UVfrq[idx], dvis.imag)
plt.plot(UVfrq[idx], sig, color="k")
plt.plot(UVfrq[idx], -sig, color="k")
plt.hlines(np.sqrt(np.mean((np.abs(dvis)**2))), 152, 168, color="r")
plt.hlines(-np.sqrt(np.mean((np.abs(dvis)**2))), 152, 168, color="r")
plt.xlabel("Frequency [MHz]")
plt.ylabel("Noise [Jy]")
```

```python
std_path = "/users/pkeller/code/H1C_IDR3.2/data/noise_std.h5"

# Noise estimated from auto-correlation visibilities
with h5py.File(std_path) as f:
    lstn = f["LST"][()]
    sig = f["RMS"][()].real
    
# interpolate NaN-values
lst_idx = np.argmin(np.abs(lstn - lstm))
lstn = lstn[lst_idx]
sig = sig[lst_idx]

# data noise estimation
idx = np.where((UVfrq > 152.25) & (UVfrq < 167.97))

dvis1 = (vis1[1] - vis1[0])[idx] / np.sqrt(2)
dvis2 = (vis2[1] - vis2[0])[idx] / np.sqrt(2)
dvis3 = (vis3[1] - vis3[0])[idx] / np.sqrt(2)
```

## Load Data

```python
trclass = "EQ14"
field = "B"
data_path = f"/lustre/aoc/projects/hera/pkeller/data/H1C_IDR3.2/sample/{trclass}_F{field}_B2.h5"

with h5py.File(data_path, "r") as f:
    freq = f["FRQ"][()]
    lst = f["LST"][()]
    eicp_xx = np.nanmean(f["eicp jdmed (2) XX"][()], axis=0)
    eicp_yy = np.nanmean(f["eicp jdmed (2) YY"][()], axis=0)
    eicp_xx_m = np.nanmean(f["eicp jdmed (2) XX model"][()], axis=0)
    eicp_yy_m = np.nanmean(f["eicp jdmed (2) YY model"][()], axis=0)
```

```python
cp_xx = np.angle(eicp_xx)
cp_yy = np.angle(eicp_yy)
cp_xx_m = np.angle(eicp_xx_m)
cp_yy_m = np.angle(eicp_yy_m)
```

```python
# ticklabels
xticks = np.array(["{:.1f}".format(f) for f in freq])
yticks = np.array(["{:.1f}".format(t) for t in lst])

fig, ax = plt.subplots(figsize=(10, 5))
im = ax.imshow(cp_xx_m[1], aspect="auto", cmap="twilight", vmin=-np.pi, vmax=np.pi)
ax.set_xlabel("Frequency [MHz]", fontsize=14)
ax.set_ylabel("LST [h]", fontsize=14)
ax.set_xticks(np.arange(0, len(freq), 20))
ax.set_yticks(np.arange(0, len(lst), 50))
ax.set_yticklabels(yticks[::50])
ax.set_xticklabels(xticks[::20])
ax.minorticks_on()
fig.colorbar(im).set_label("Closure Phase [rad]", size=14)
plt.show()
```

```python
cp = cp_xx

def plot_spectrogram(tr):
    fig, ax = plt.subplots(figsize=(10, 5))
    im = ax.imshow(cp[tr], aspect="auto", cmap="twilight", vmin=-np.pi, vmax=np.pi)
    ax.set_xlabel("Frequency [MHz]", fontsize=14)
    ax.set_ylabel("LST [h]", fontsize=14)
    ax.set_xticks(np.arange(0, len(freq), 20))
    ax.set_yticks(np.arange(0, len(lst), 50))
    ax.set_yticklabels(yticks[::50])
    ax.set_xticklabels(xticks[::20])
    ax.minorticks_on()
    fig.colorbar(im).set_label("Closure Phase [rad]", size=14)
    plt.show()

interact(plot_spectrogram, tr=(0, cp.shape[0]-1))
```

```python
cp = cp_xx
cp_m = cp_xx_m

def plot_spectrum(tr, t):
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.scatter(freq, cp[tr, t], label="data")
    ax.scatter(freq, cp_m[tr, t], label="model")
    ax.set_xlabel("Frequency [MHz]", fontsize=14)
    ax.set_ylabel("Closure Phase [rad]", fontsize=14)
    plt.ylim([-np.pi, np.pi])
    plt.legend()
    plt.show()

interact(plot_spectrum, tr=(0, cp.shape[0]-1), t=(0, cp.shape[1]-1))
```
