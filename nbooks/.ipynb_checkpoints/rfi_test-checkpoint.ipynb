```python
import imp; imp.reload(rfi)
N = 120 

# median filter along frequency axis
for t in range(RFIf.shape[0]):
    eicp  = np.exp(1j * RFIf[t])
    flags[t] += rfi.moving_median_clip(eicp, 10, 7)
RFIf[flags] = np.nan

# median filter along time axis
for f in range(RFIf.shape[1]):
    eicp = np.exp(1j * RFIf[:, f])
    flags[:, f] += rfi.moving_median_clip(eicp, 12, 7)
RFIf[flags] = np.nan

# polynomial fitting along frequency axis
for t in range(RFIf.shape[0]):
    eicp = np.exp(1j * np.angle(RFIf[t]))
    flags[t] += rfi.poly_clip(eicp, frq, 6, 16)
RFIf[flags] = np.nan
    
# polynomial fitting along time axis
for f in range(RFIf.shape[1]):
    eicp = np.exp(1j * np.angle(RFIf[:, f]))
    flags[:, f] += rfi.poly_clip_slide(eicp, lst, N, 6, 10)
RFIf[flags] = np.nan  

# polynomial fitting along frequency axis
for t in range(RFIf.shape[0]):
    flags[t] += rfi.poly_clip(np.abs(RFIf[t]), frq, 6, 16)
RFIf[flags] = RFIf[flags] * np.nan

# polynomial fitting along time axis
for f in range(RFIf.shape[1]):
    flags[:, f] += rfi.poly_clip_slide(np.abs(RFIf[:, f]), lst, N, 6, 10)
RFIf[flags] = RFIf[flags] * np.nan

# polynomial fitting along frequency axis
for t in range(RFIf.shape[0]):
    flags[t] += rfi.poly_clip(RFIf[t], frq, 6, 16)

# polynomial fitting along time axis
for f in range(RFIf.shape[1]):
    flags[:, f] += rfi.poly_clip_slide(RFIf[:, f], lst, N, 6, 10)
    
# flag adjacent data points
for i in range(RFIf.shape[0] // N):
    s = slice(i*N, (i+1)*N)
    flags[s] = rfi.poly_clip_adj_2D(np.abs(RFIf[s]), lst[s], frq, n=5, deg=10)
flags[-N:] = rfi.poly_clip_adj_2D(np.abs(RFIf[-N:]), lst[-N:], frq, n=5, deg=10)
RFIf[flags] = RFIf[flags] * np.nan

# flag adjacent data points
for i in range(RFIf.shape[0] // N):
    s = slice(i*N, (i+1)*N)
    flags[s] = rfi.poly_clip_adj_2D(RFIf[s], lst[s], frq, n=5, deg=10)
flags[-N:] = rfi.poly_clip_adj_2D(RFIf[-N:], lst[-N:], frq, n=5, deg=10)
```

## Plot Time Slice

```python
def plot(t):
    print(f"{lst[t]} h")
    
    idx = ~np.isnan(RFIf[t])
    nsig = 3
    
    eicp = np.exp(1j * np.angle(RFIf[t]))
    fit = np.exp(1j * np.angle(model[t][idx]))
     
    res = eicp[idx] - fit
    std = np.nanstd(res)
        
    fig, ax = plt.subplots(1, 2, figsize=(15, 7), sharey=True)
    ax[0].scatter(frq[idx], eicp[idx].real, s=0.5)
    ax[1].scatter(frq[idx], eicp[idx].imag, s=0.5)
    ax[0].plot(frq[idx], fit.real)
    ax[1].plot(frq[idx], fit.imag)
    ax[0].plot(frq[idx], (fit.real + nsig * std), color="orange")
    ax[1].plot(frq[idx], (fit.imag + nsig * std), color="orange")
    ax[0].plot(frq[idx], (fit.real - nsig * std), color="orange")
    ax[1].plot(frq[idx], (fit.imag - nsig * std), color="orange")
    ax[0].scatter(frq[~idx], np.exp(1j * np.angle(RFI[t][~idx])).real, s=10, color="r")
    ax[1].scatter(frq[~idx], np.exp(1j * np.angle(RFI[t][~idx])).imag, s=10, color="r")
    
    ax[0].set_title(f"JD {jd}, Real")
    ax[1].set_title(f"JD {jd}, Imaginary")
    plt.setp(ax, xlabel="Frequency (MHz)", ylabel="Complex Phase", xlim=[min(frq), max(frq)], ylim=[1, -1])
    
    plt.tight_layout()
    plt.show()

interact(plot, t=(0, len(lst)-1))
```

## Plot Frequency Slice

```python
n = 120

def plot(t, f):
    print(f"{frq[f]} MHz")
    s = slice(t*n, (t+1)*n)
    
    idx = ~np.isnan(RFIf[s, f])
    nsig = 6
    
    eicp = np.exp(1j * np.angle(RFIf[s, f]))
    fit = np.exp(1j * np.angle(model[s, f]))
    
    
    res = eicp[idx] - fit
    std = np.nanstd(res)
        
    fig, ax = plt.subplots(1, 2, figsize=(15, 7), sharey=True)
    ax[0].scatter(lst[s], eicp.real, s=0.5)
    ax[1].scatter(lst[s], eicp.imag, s=0.5)
    ax[0].plot(lst[s][idx], fit.real)
    ax[1].plot(lst[s][idx], fit.imag)
    ax[0].plot(lst[s][idx], (fit.real + nsig * std), color="orange")
    ax[1].plot(lst[s][idx], (fit.imag + nsig * std), color="orange")
    ax[0].plot(lst[s][idx], (fit.real - nsig * std), color="orange")
    ax[1].plot(lst[s][idx], (fit.imag - nsig * std), color="orange")
    ax[0].scatter(lst[s][~idx], np.exp(1j * np.angle(RFI[s, f][~idx])).real, s=10, color="r")
    ax[1].scatter(lst[s][~idx], np.exp(1j * np.angle(RFI[s, f][~idx])).imag, s=10, color="r")
    
    ax[0].set_title(f"JD {jd}, Real")
    ax[1].set_title(f"JD {jd}, Imaginary")
    plt.setp(ax, xlabel="LST (h)", ylabel="Complex Phase", xlim=[min(lst[s]), max(lst[s])], ylim=[-1, 1])
    
    plt.tight_layout()
    plt.show()

interact(plot, t=(0, RFIf.shape[0] // n), f=(0, len(frq)-1))
```
