# Averaging Triads 

```python
import os
import h5py
import numpy as np
from scipy.signal.windows import get_window

import matplotlib.pyplot as plt

import sys
sys.path.append("/users/pkeller/code/ClosurePhaseAnalysis/")

from library import dspec
from library import cptools as cp
```

## Load Data

```python
# data path
path = "/lustre/aoc/projects/hera/pkeller/data/H1C_IDR3.2/sample/EQ14_FC_B2_AVG.h5"
xps_path = "/lustre/aoc/projects/hera/pkeller/data/H1C_IDR3.2/sample/EQ14_FC_B2_XPS.h5"

# load covariance data
with h5py.File(path, "r") as f:
    frq = f["FRQ"][()]
    lst = f["LST"][()]
    cov_xx = f["cov XX"][()]
    cov_yy = f["cov YY"][()]
    eicp_xx = f["eicp XX 1"][()]
    eicp_yy = f["eicp YY 1"][()]
    
# load cross-power spectral data
with h5py.File(xps_path, "r") as f:
    xps_xx = f["XPS XX"][()]
    xps_yy = f["XPS YY"][()]
    
Nlst = len(lst)
Ntrx = eicp_xx.shape[1]
Ntry = eicp_yy.shape[1]
Ndx = eicp_xx.shape[-1]
Ndy = eicp_yy.shape[-1]
```

## Sums and Differences
The differences allow to investigate how different averaging methods affect the noise levels.

```python
sum_xx = (eicp_xx[0] + eicp_xx[1]) / 2
dif_xx = (eicp_xx[0] - eicp_xx[1]) / 2
sum_yy = (eicp_yy[0] + eicp_yy[1]) / 2
dif_yy = (eicp_yy[0] - eicp_yy[1]) / 2
```

## Invert Covariance Matrices and Compute Weights
The minimum variance weights for dependent variables are $w_j=\sum_{i} C^{-1}_{ij}$, where $C_{ij}$ is the covariance between variables $x_i$ and $x_j$. 

```python
cov_xx_inv = np.linalg.inv(cov_xx.real)
cov_yy_inv = np.linalg.inv(cov_xx.real)

# covariance weights
w_cov_xx = np.sum(cov_xx_inv, axis=1).swapaxes(0, 1)
w_cov_yy = np.sum(cov_yy_inv, axis=1).swapaxes(0, 1)

# variance weights
w_var_xx = np.array([1 / (cov_xx[:, i, i]) for i in range(Ntrx)])
w_var_yy = np.array([1 / (cov_yy[:, i, i]) for i in range(Ntry)])
```

## Delay Transform

```python
# sampling frequency
fs = 10.24 * 1e-6

# delay array
delay = dspec.delay(n=Ndx)

# window function
w = get_window("blackmanharris", Ndx)

# delay transform sums and differences
sum_xx_ft = dspec.fft(sum_xx, W=w, shift=True) / fs
sum_yy_ft = dspec.fft(sum_yy, W=w, shift=True) / fs
dif_xx_ft = dspec.fft(dif_xx, W=w, shift=True) / fs
dif_yy_ft = dspec.fft(dif_yy, W=w, shift=True) / fs
```

```python
cov = np.zeros((Nlst, Ntrx, Ntrx, Ntrx, Ntrx))
for i, j, k, l in product(range(Ntrx), range(Ntrx), range(Ntrx), range(Ntrx)):
    cov[:, i, j, k, l] = cov_xx[:, i, k].real * cov_xx[:, j, l].real
```

```python
cov_inv = np.linalg.inv(cov)
w_cov_xx = cov_inv.sum(axis=1).reshape(Nlst, Ntrx, Ntrx)
```

```python
w_cov_xx = np.moveaxis(w_cov_xx, 0, -1)
```

```python
w_cov_xx = np.sum(cov_xx_inv, axis=1).swapaxes(0, 1)
w_var_xx = np.array([1 / (cov_xx[:, i, i]) for i in range(Ntrx)])
```

```python
sum_mean = np.mean(eicp_xx_sum, axis=0)
sum_median = cp.geomed(eicp_xx_sum, axis=0)
sum_cov = ((np.moveaxis(eicp_xx_sum, -1, 0) * w_cov_xx).sum(axis=1) / w_cov_xx.sum(axis=0)).swapaxes(0, 1)
sum_var = ((np.moveaxis(eicp_xx_sum, -1, 0) * w_var_xx).sum(axis=1) / w_var_xx.sum(axis=0)).swapaxes(0, 1)
```

```python
dif_mean = np.mean(eicp_xx_dif, axis=0)
dif_median = cp.geomed(eicp_xx_dif, axis=0)
dif_cov = ((np.moveaxis(eicp_xx_dif, -1, 0) * w_cov_xx).sum(axis=1) / w_cov_xx.sum(axis=0)).swapaxes(0, 1)
dif_var = ((np.moveaxis(eicp_xx_dif, -1, 0) * w_var_xx).sum(axis=1) / w_var_xx.sum(axis=0)).swapaxes(0, 1)
```

```python
fs = 10.24 * 1e-6
delay = dspec.delay(n=Ndx)
w = get_window("blackmanharris", Ndx)

sum_ps_mean = dspec.xpspec(sum_mean, sum_mean, W=w, fs=fs, shift=True)
sum_ps_median = dspec.xpspec(sum_median, sum_median, W=w, fs=fs, shift=True)
sum_ps_var = dspec.xpspec(sum_var, sum_var, W=w, fs=fs, shift=True)
sum_ps_cov = dspec.xpspec(sum_cov, sum_cov, W=w, fs=fs, shift=True)

dif_ps_mean = dspec.xpspec(dif_mean, dif_mean, W=w, fs=fs, shift=True)
dif_ps_median = dspec.xpspec(dif_median, dif_median, W=w, fs=fs, shift=True)
dif_ps_var = dspec.xpspec(dif_var, dif_var, W=w, fs=fs, shift=True)
dif_ps_cov = dspec.xpspec(dif_cov, dif_cov, W=w, fs=fs, shift=True)
```

```python
sum_ft = dspec.fft(eicp_xx_sum, W=w, shift=True) / fs
sum_ft_median = np.array([cp.geomed(sum_ft[:, i], weights=w_var_xx[:, i], axis=0) for i in range(Nlst)])
sum_ps_median = sum_ft_median * sum_ft_median.conjugate()
```
